```python
from scipy import stats

def Frecuencia_Relativa_Delito_Municipio(dfDelictiva, dfClima, municipio, delito, factor_climatologico):

    dfDelictivaMunicipio = dfDelictiva[dfDelictiva['municipio'] == municipio ].groupby(['anio','mes','tipo_delito'])[['numero_delitos']].agg('sum')
    dfDelictivaMunicipio.reset_index(inplace = True)

    dfClimaMunicipio = dfClima[dfClima['municipio'] == municipio]
    dfDelictivaMunicipioDelito = dfDelictivaMunicipio[dfDelictivaMunicipio['tipo_delito'] == delito]
    MergeRegion = dfDelictivaMunicipioDelito.merge(dfClimaMunicipio, how = 'left', on = ['anio','mes'])
    MergeRegion.head()

    num_total = dfDelictivaMunicipioDelito.groupby('anio')['numero_delitos'].agg('sum')
    num_total

    MergeRelReg = MergeRegion.merge(num_total, how = 'left', on = 'anio')
    MergeRelReg.head()

    MergeRelReg['numero_delito_rel'] = MergeRelReg['numero_delitos_x']/MergeRelReg['numero_delitos_y']
    MergeRelRegN = MergeRelReg[MergeRelReg['anio']<2023]

    plt.scatter(x = MergeRelRegN[factor_climatologico], y = MergeRelRegN['numero_delito_rel'], s = 5)
    plt.show()

    return [np.corrcoef(MergeRelRegN[factor_climatologico], MergeRelRegN['numero_delito_rel'])[0][1], 
            stats.spearmanr(MergeRelRegN[factor_climatologico],MergeRelRegN['numero_delito_rel'])]
```

```python
import plotly.express as px

def Serie_Tiempo(dfDelictiva, delito):

    DiccionarioMeses = {
    'Enero':1,
    'Febrero':2,
    'Marzo':3,
    'Abril':4,
    'Mayo':5,
    'Junio':6,
    'Julio':7,
    'Agosto':8,
    'Septiembre':9,
    'Octubre':10,
    'Noviembre':11,
    'Diciembre':12
    }

    DiccionarioAnios = {
        2015:1,
        2016:2,
        2017:3,
        2018:4,
        2019:5,
        2020:6,
        2021:7,
        2022:8,
        2023:9
    }

    dfDelictivaEstatal = dfDelictiva.groupby(['anio','mes','tipo_delito'])[['numero_delitos']].agg('sum')
    dfDelictivaEstatal.reset_index(inplace = True)
    dfDelictivaEstatalDelito = dfDelictivaEstatal[dfDelictivaEstatal['tipo_delito'] == delito]

    dfDelictivaEstatalDelito['periodo'] = dfDelictivaEstatalDelito['anio'].map(lambda x: 12*(DiccionarioAnios[x]-1)) + dfDelictivaEstatalDelito['mes'].map(lambda x: DiccionarioMeses[x])
    dfDelictivaEstatalDelito = dfDelictivaEstatalDelito.sort_values('periodo')
    fig = px.line(dfDelictivaEstatalDelito, x = 'periodo', y = 'numero_delitos', title = f'{delito} en Sonora')
    fig.show()
```

```python
def Serie_Tiempo_Delitos(dfEstatal, delitos):

    dfGrouped = dfEstatal.groupby(['anio','mes','tipo_delito'])['numero_delitos'].agg('sum').reset_index()
    dfGrouped = dfGrouped[dfGrouped['tipo_delito'].isin(delitos)]
    dfGrouped = dfGrouped.set_index(['anio','mes'])

    dfPivot = dfGrouped.pivot(columns = 'tipo_delito', values = 'numero_delitos')
    dfPivot.reset_index(inplace = True)

    dfPivot['periodo'] = dfPivot['anio'].map(lambda x: 12*(DiccionarioAnios[x]-1)) + dfPivot['mes'].map(lambda x: DiccionarioMeses[x])
    dfPivot.sort_values('periodo', inplace = True)

    fig = px.scatter(dfPivot, x = 'periodo', y = delitos)
    fig.show()
```

```python
def Scatter_Delito_Factor_Climatologico_Municipio(dfDelictiva, dfClima, municipio, delitos, factor_climatologico):
    dfDelictivaMunicipio = dfDelictiva[dfDelictiva['municipio'] == municipio ].groupby(['anio','mes','tipo_delito'])[['numero_delitos']].agg('sum')
    dfDelictivaMunicipio.reset_index(inplace = True)

    dfClimaMunicipio = dfClima[dfClima['municipio'] == municipio]
    dfDelictivaMunicipioDelito = dfDelictivaMunicipio[dfDelictivaMunicipio['tipo_delito'].isin(delitos)]
    MergeRegion = dfDelictivaMunicipioDelito.merge(dfClimaMunicipio, how = 'left', on = ['anio','mes'])
    MergeRegion.head()

    num_total = dfDelictivaMunicipioDelito.groupby(['anio','tipo_delito'])['numero_delitos'].agg('sum')
    num_total

    MergeRelReg = MergeRegion.merge(num_total, how = 'left', on = ['anio','tipo_delito'])
    MergeRelReg['numero_delito_rel'] = MergeRelReg['numero_delitos_x']/MergeRelReg['numero_delitos_y']
    MergeRelRegN = MergeRelReg[MergeRelReg['anio']<2023]
    MergeRelRegN.set_index(['anio','mes',factor_climatologico], inplace = True)

    dfPivot = MergeRelRegN.pivot(columns = 'tipo_delito', values = 'numero_delito_rel')
    dfPivot.reset_index(inplace = True)

    fig = px.scatter(dfPivot, x = factor_climatologico, y = delitos)
    fig.show()

    for delito in delitos:
        print(f'{delito} en {municipio}:' + str([np.corrcoef(dfPivot[factor_climatologico], dfPivot[delito])[0][1], 
            stats.spearmanr(dfPivot[factor_climatologico],dfPivot[delito])]))
        print('\n')
```

```python
def Scatter_Delito_Factor_Climatologico_Local(dfDelictiva, dfClima, local, delitos, factor_climatologico, escala = 'region'):

    if(escala == 'region'):
        dfDelictivaLocal = dfDelictiva[dfDelictiva['region'] == local].groupby(['anio','mes','tipo_delito'])[['numero_delitos']].agg('sum')
        dfDelictivaLocal.reset_index(inplace = True)
        dfClimaLocal = dfClima[dfClima['region'] == local]
    else:
        dfDelictivaLocal = dfDelictiva[dfDelictiva['municipio'] == local].groupby(['anio','mes','tipo_delito'])[['numero_delitos']].agg('sum')
        dfDelictivaLocal.reset_index(inplace = True)
        dfClimaLocal = dfClima[dfClima['municipio'] == local]
        
    dfDelictivaLocalDelito = dfDelictivaLocal[dfDelictivaLocal['tipo_delito'].isin(delitos)]
    MergeLocal = dfDelictivaLocalDelito.merge(dfClimaLocal, how = 'left', on = ['anio','mes'])
    MergeLocal.head()

    num_total = dfDelictivaLocalDelito.groupby(['anio','tipo_delito'])['numero_delitos'].agg('sum')
    num_total

    MergeRelLoc = MergeLocal.merge(num_total, how = 'left', on = ['anio','tipo_delito'])
    MergeRelLoc['numero_delito_rel'] = MergeRelLoc['numero_delitos_x']/MergeRelLoc['numero_delitos_y']
    MergeRelLocN = MergeRelLoc[MergeRelLoc['anio']<2023]
    MergeRelLocN.set_index(['anio','mes',factor_climatologico], inplace = True)

    dfPivot = MergeRelLocN.pivot(columns = 'tipo_delito', values = 'numero_delito_rel')
    dfPivot.reset_index(inplace = True)

    fig = px.scatter(dfPivot, x = factor_climatologico, y = delitos)
    fig.show()

    for delito in delitos:
        print(f'{delito} en {local}:' + str([np.corrcoef(dfPivot[factor_climatologico], dfPivot[delito])[0][1], 
            stats.spearmanr(dfPivot[factor_climatologico],dfPivot[delito])]))
        print('\n')
```
